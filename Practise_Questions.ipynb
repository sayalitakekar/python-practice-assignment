14.	Digitalize the book allotment process for school. Charges are mentioned here in the given table:  


```python


books = {
    'Hindi':   {'1-4': 60, '5-8': 100, '9-10': 150},
    'Marathi': {'1-4': 60, '5-8': 100, '9-10': 150},
    'English': {'1-4': 80, '5-8': 100, '9-10': 150},
    'Science': {'1-4': 90, '5-8': 120, '9-10': 200},
    'Maths':   {'1-4': 100, '5-8': 140, '9-10': 250}
}

notebooks = {
    'square':        {100: 40, 200: 70},
    '4lines':        {100: 30, 200: 50},
    '2lines':        {100: 30, 200: 50},
    'single lines':  {100: 60, 200: 100},
    'A4 notebook':   {100: 100, 200: 180}
}


def find_standard_group(std):
    if 1 <= std <= 4:
        return '1-4'
    elif 5 <= std <= 8:
        return '5-8'
    elif 9 <= std <= 10:
        return '9-10'
    else:
        return None


try:
    total_all_students = 0

    num_students = int(input("Enter number of students buying together: "))

    for i in range(num_students):
        print(f"\n===== Student {i+1} Details =====")
        std = int(input("Enter student standard (1-10): "))
        group = find_standard_group(std)

        if not group:
            print("Invalid standard entered. Skipping this student.")
            continue

        print(f"\nAvailable Books: {list(books.keys())}")
        selected_books = input("Enter subjects separated by comma: ").split(',')
        selected_books = [b.strip().capitalize() for b in selected_books]

        total_book_cost = 0
        print("\n Book Charges:")
        for book in selected_books:
            if book in books:
                price = books[book][group]
                total_book_cost += price
                print(f"{book}: ₹{price}")
            else:
                print(f"{book} not found!")

       
        print("\n Notebook Types:", list(notebooks.keys()))
        notebook_type = input("Enter notebook type: ").strip().lower()
        pages = int(input("Enter pages (100/200): "))

        if notebook_type in notebooks and pages in notebooks[notebook_type]:
            notebook_price = notebooks[notebook_type][pages]
            qty = int(input("Enter quantity of notebooks: "))
            total_notebook_cost = notebook_price * qty
            print(f"{notebook_type.title()} Notebook ({pages} pages): ₹{notebook_price} x {qty} = ₹{total_notebook_cost}")
        else:
            total_notebook_cost = 0
            print("Invalid notebook choice or page count!")

        student_total = total_book_cost + total_notebook_cost
        total_all_students += student_total

        print("\n------------------------------------")
        print(f"Student {i+1} Book Cost: ₹{total_book_cost}")
        print(f"Student {i+1} Notebook Cost: ₹{total_notebook_cost}")
        print(f"Student {i+1} Total: ₹{student_total}")
        print("------------------------------------")


    print("\n FINAL BILL (All Students)")
    print("====================================")
    print(f"Combined Total for {num_students} students: ₹{total_all_students}")
    print("====================================")

except ValueError:
    print(" Please enter numeric values only where required.")
except Exception as e:
    print(" Something went wrong:", e)
```

15.	Create a interest bucket for Fix Deposit in the bank. Ask user to enter start date and end date for the FD and check which bucket list it belongs to and assign the interest rate


```python
from datetime import datetime

# Define interest buckets
interest_buckets = [
    (7, 45, 5.75, 6.25),
    (46, 179, 6.25, 6.75),
    (180, 210, 6.35, 6.85),
    (211, 364, 6.40, 6.90),
    (365, 729, 7.00, 7.50),
    (730, 1094, 6.75, 7.25),
    (1095, 1824, 6.70, 7.20),
    (1825, 3650, 6.60, 7.10),
]

def get_interest_rate(start_date, end_date, is_senior=False):
    # Convert string to date
    s = datetime.strptime(start_date, "%d-%m-%Y")
    e = datetime.strptime(end_date, "%d-%m-%Y")

    # Calculate number of days
    days = (e - s).days

    # Find correct interest bucket
    for low, high, public_rate, senior_rate in interest_buckets:
        if low <= days <= high:
            rate = senior_rate if is_senior else public_rate
            return {
                "days": days,
                "bucket": f"{low}-{high} days",
                "interest_rate": rate
            }

    return {"days": days, "bucket": "Out of range", "interest_rate": None}

# Example usage
start = input("Enter FD start date (dd-mm-yyyy): ")
end = input("Enter FD end date (dd-mm-yyyy): ")
senior = input("Are you a senior citizen? (Y/N): ").strip().upper() == "Y"

result = get_interest_rate(start, end, senior)

print("\n📄 Fixed Deposit Details")
print(f"Duration: {result['days']} days")
print(f"Tenor Bucket: {result['bucket']}")
if result['interest_rate']:
    print(f"Applicable Interest Rate: {result['interest_rate']}%")
else:
    print("No matching interest bucket found.")
```

### 27. What is the difference between add and update methods in set? 

They both add elements, but the way they add is very different
1️ add() Method
set_name.add(element)

Adds a single element to the set.
The element can be number, string, tuple, etc. (immutable types only).
If the element already exists, the set remains unchanged (no duplicates).

example:
my_set = {1, 2, 3}
my_set.add(4)
print(my_set)

2️ update() Method

set_name.update(iterable)

Adds multiple elements (from another iterable like list, tuple, string, or another set).
It iterates through the given iterable and adds each element one by one.

example:
my_set = {1, 2, 3}
my_set.update([4, 5])
print(my_set)


#### 28. What is the difference between append and extend methods in list?

##### 1️ append() Method
list_name.append(element)

Adds one single element at the end of the list.
The element can be a number, string, list, tuple, etc.
If you append a list, it gets added as a single item (nested list).


##### extend() Method
list_name.extend(iterable)

Adds multiple elements to the list (takes an iterable like another list, tuple, or string).
It goes through each element of the iterable and adds them one by one to the original list.



```python
fruits = ["apple", "banana"]
fruits.append("mango")
print(fruits)
```

```python
fruits.append(["grape", "kiwi"])
print(fruits)
```

### 29.	What is the difference between pop and remove methods? 

### 1️ remove() Method
Syntax:
list_name.remove(value)

Description:
Removes the first matching value from the list.
You must specify the element’s value, not its index.
If the value is not present, it raises a ValueError.

Example:
numbers = [10, 20, 30, 20, 40]
numbers.remove(20)
print(numbers)


Output:
[10, 30, 20, 40]
It removes the first occurrence of 20.

If the element doesn’t exist:
numbers.remove(100)

Output:
ValueError: list.remove(x): x not in list

 #### 2️ pop() Method
Syntax:
list_name.pop(index)

Description:
Removes and returns the element at the given index position.
If no index is given, it removes the last element.
If the index is out of range, it raises an IndexError.

Example:
numbers = [10, 20, 30, 40]
numbers.pop(2)
print(numbers)


Output:
[10, 20, 40]


Removes the element at index 2 (which is 30).

If no index is provided:
numbers.pop()
print(numbers)


Output
[10, 20]

### 30.	What is the difference between discard, pop, remove methods? 
#### 1️ remove() Method
Syntax:
set_name.remove(element)

Description:

Removes the specified element from the set.
If the element is not found, it raises a KeyError.
It removes only one specific element.

Example:
my_set = {10, 20, 30, 40}
my_set.remove(20)
print(my_set)


Output:

{10, 30, 40}


If the element doesn’t exist:
my_set.remove(100)

Error:
KeyError: 100

### 2️ discard() Method
Syntax:
set_name.discard(element)

Description:

Also removes a specific element from the set.
BUT — if the element doesn’t exist, it does NOT raise an error (safe removal).

 Example:
my_set = {10, 20, 30, 40}
my_set.discard(20)
print(my_set)

#### 54.	Create a table  cust_info as sr_no, name, DOB, mobile. Ask user to enter the information from python code. Validate all fields and after validation insert records in the table.

```python
import sqlite3
import re

conn = sqlite3.connect("customer.db")
cursor = conn.cursor()

cursor.execute("""
CREATE TABLE IF NOT EXISTS cust_info (
    sr_no INTEGER PRIMARY KEY AUTOINCREMENT,
    name TEXT NOT NULL,
    dob TEXT NOT NULL,
    mobile TEXT NOT NULL
)
""")

print(" Table 'cust_info' ready.\n")


def validate_name(name):
    if not name.isalpha():
        raise ValueError(" Name must contain only alphabets (no numbers or symbols).")
    return name

def validate_dob(dob):
    pattern = r'^\d{2}-\d{2}-\d{4}$'
    if not re.match(pattern, dob):
        raise ValueError(" DOB must be in format dd-mm-yyyy.")
    return dob

def validate_mobile(mobile):
    if not (mobile.isdigit() and len(mobile) == 10):
        raise ValueError(" Mobile must be exactly 10 digits (numbers only).")
    return mobile


try:
    name = validate_name(input("Enter Name: "))
    dob = validate_dob(input("Enter DOB (dd-mm-yyyy): "))
    mobile = validate_mobile(input("Enter Mobile Number: "))

    cursor.execute("INSERT INTO cust_info (name, dob, mobile) VALUES (?, ?, ?)", (name, dob, mobile))
    conn.commit()
    print("\n Record inserted successfully!")

except ValueError as e:
    print(e)
except Exception as e:
    print(" Unexpected Error:", e)
finally:
    conn.close()
```
